```python
import requests
import bs4
```

```python
'http://books.toscrape.com/catalogue/page-2.html'
```

```python
base_url='http://books.toscrape.com/catalogue/page-{}.html'
```

```python
res= requests.get(base_url.format(1))
```

```python
soup =bs4.BeautifulSoup(res.text,'lxml')
```

```python
products=soup.select(".product_pod")
```

```python
len(soup.select(".product_pod"))
```

```python
example =products[0]
example
```

```python
'star-rating Three' in str(example)
```

```python
example.select(".star-rating.Three")
```

```python
example.select('a')[1]['title']
```

```python
#We can check if something is 2 stars (string call in, example.select(rating))
# example.select('a')[1]['title'] to grab the book title
```

```python
two_star_titles =[]
for n in range(1,51):
    scarpe_url=base_url.format(n)
    res = requests.get(scarpe_url)
    
    soup=bs4.BeautifulSoup(res.text,'lxml')
    books=soup.select(".product_pod")
    
    for book in books:
        if len(book.select('.star-rating.Two')) !=0:
            book_title = book.select('a')[1]['title']
            two_star_titles.append(book_title)
```
